```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以获得可重现的结果
np.random.seed(42)

# 生成第一象限的黄色点
# 使用均匀分布在0到1之间生成x和y的值
num_points = 100
x_yellow = np.random.uniform(0.5, 1.0, num_points)
y_yellow = np.random.uniform(0.5, 1.0, num_points)

# 生成第二象限的蓝色点
x_blue = np.random.uniform(0.0, 0.5, num_points)
y_blue = np.random.uniform(0.5, 1.0, num_points)

# 生成第三象限的黄色点
x_yellow_3 = np.random.uniform(0.0, 0.5, num_points)
y_yellow_3 = np.random.uniform(0.0, 0.5, num_points)

# 生成第四象限的蓝色点
x_blue_4 = np.random.uniform(0.5, 1.0, num_points)
y_blue_4 = np.random.uniform(0.0, 0.5, num_points)

# 绘制散点图
plt.scatter(x_yellow, y_yellow, color='yellow', edgecolor='k', label='Yellow Points (1st Quadrant)')
plt.scatter(x_blue, y_blue, color='blue', edgecolor='k', label='Blue Points (2nd Quadrant)')
plt.scatter(x_yellow_3, y_yellow_3, color='yellow', edgecolor='k', label='Yellow Points (3rd Quadrant)')
plt.scatter(x_blue_4, y_blue_4, color='blue', edgecolor='k', label='Blue Points (4th Quadrant)')

# 设置坐标轴范围
plt.xlim(0, 1)
plt.ylim(0, 1)

# 添加图例
plt.legend()

# 添加坐标轴标签和标题
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Quadrant Points')

# 显示网格
plt.grid(True)

# 显示图形
plt.show()


# 生成所有点的位置数组
data = np.vstack((np.column_stack((x_yellow, y_yellow)),
                    np.column_stack((x_blue, y_blue)),
                    np.column_stack((x_yellow_3, y_yellow_3)),
                    np.column_stack((x_blue_4, y_blue_4))))

# 生成标签数组，1和3象限为黄色（标签0），2和4象限为蓝色（标签1）
labels = np.array([0]*num_points + [1]*num_points + [0]*num_points + [1]*num_points)

# 确认数组形状
print("Shape of points:", data.shape)
print("Shape of labels:", labels.shape)

# 返回结果
data, labels

```

```python
plt.scatter(data[:,0][labels==1],data[:,1][labels==1])
plt.scatter(data[:,0][labels==0],data[:,1][labels==0])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data,labels,test_size=0.33,random_state=10)
print(x_train.shape,x_test.shape,data.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

mlp = Sequential()
mlp.add(Dense(units=20, input_dim =2, activation = 'sigmoid'))
mlp.add(Dense(units=1, activation ='sigmoid')) 
mlp.summary()
```

```python
mlp.compile(optimizer = 'adam', loss = 'binary_crossentropy')
```

```python
mlp.fit(x_train,y_train,epochs=3000)
```

```python
y_train_pred_probs = mlp.predict(x_train)  # 这将返回形状为 (n_samples, 1) 的概率数组
y_train_pre = (y_train_pred_probs >= 0.5).astype(int)  # 将概率转换为类别标签

y_test_pred_probs = mlp.predict(x_test)  # 这将返回形状为 (n_samples, 1) 的概率数组
y_test_pre = (y_test_pred_probs >= 0.5).astype(int)  # 将概率转换为类别标签

from sklearn.metrics import accuracy_score
acc_train = accuracy_score(y_train,y_train_pre)
acc_test = accuracy_score(y_test,y_test_pre)
print('train accuracy_score :',acc_train)
print('test accuracy_score :',acc_test)
```
